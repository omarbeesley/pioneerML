# Pion Stop Regressor
Estimate pion stop position from standardized group features.

```
import torch
import torch.nn.functional as F
from torch_geometric.loader import DataLoader
from graph_data.utils import GraphGroupDataset, GraphRecord
from graph_data.models import PionStopRegressor
```

```
def build_records(groups):
    records = []
    for group in groups:
        records.append(GraphRecord(
            coord=group[:, 0],
            z=group[:, 1],
            energy=group[:, 2],
            view=group[:, 3],
        ))
    return records
```
